```python
df=pd.read_csv("employee_promotion.csv")
plt.plot(df.department)
plt.show()
```

```python
df[df.department].count()
```

```python
ged=df.groupby('age')['avg_training_score'].mean()
plt.bar(ged.index,ged.values,color='red')
plt.title('age vs average training score')
plt.xlabel('age')
plt.ylabel('average training score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.plot(df['age'],df['avg_training_score'],color='purple',marker='o')
plt.title('AGE VS AVERAGE TRAINING SCORE (LINE PLOT)')
plt.xlabel('Age')
plt.ylabel('Average Training Score')
plt.show()
```

```python
import seaborn as sns
de = pd.read_csv("employee_promotion.csv", usecols=['no_of_trainings', 'age', 'previous_year_rating',
       'length_of_service', 'awards_won', 'avg_training_score', 'is_promoted'])
corr_matrix = de.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
ged=df.groupby('age')['avg_training_score'].mean()
plt.pie(ged,labels=ged.index)
plt.title('average training score (pie chart)')
plt.show()
```
