```python
import matplotlib.pyplot as plt  
x = [1, 2, 3, 4] 
e  =(0.1, 0, 0, 0)
plt.pie(x,e)
plt.title("Pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt  
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
plt.scatter(x, y)
plt.legend("A")
plt.title("Scatter chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
Studying = [7, 8, 6, 11, 7]
playing = [8, 5, 7, 8, 13]
plt.stackplot(days, Studying, playing,
            colors =['r', 'c'])
plt.xlabel('Days')
plt.ylabel('No of Hours')
plt.title('Representation of Study and \
Playing wrt to Days')
plt.show()
```
